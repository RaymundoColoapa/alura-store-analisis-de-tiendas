```python
# Importar pandas
import pandas as pd

# Cargar los datos desde las URLs oficiales del desafío
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer cada tienda en un DataFrame individual
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

#  columna "Tienda" a cada DataFrame
tienda1["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"

# Unir todos los DataFrames en uno solo
df_completo = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

# Crear columna "Ingresos" (Precio - Costo de envío)
df_completo["Ingresos"] = df_completo["Precio"] - df_completo["Costo de envío"]

# Mostrar los primeros registros para verificar
display(df_completo.head())
```

```python
# Agrupar por tienda y calcular los ingresos totales
facturacion_por_tienda = df_completo.groupby("Tienda")["Ingresos"].sum().reset_index()

# Ordenar de mayor a menor facturación
facturacion_por_tienda = facturacion_por_tienda.sort_values(by="Ingresos", ascending=False)

# Mostrar resultados
print("Facturación total por tienda:")
display(facturacion_por_tienda)
```

```python
import matplotlib.pyplot as plt

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(facturacion_por_tienda["Tienda"], facturacion_por_tienda["Ingresos"], color='skyblue')

# Títulos y etiquetas
plt.title("Facturación Total por Tienda", fontsize=14)
plt.xlabel("Tienda", fontsize=12)
plt.ylabel("Ingresos (MXN)", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
print(df_completo.columns)

```

```python
# Contar cuántas ventas hubo por categoría (cantidad de filas por categoría)
ventas_por_categoria = df_completo["Categoría del Producto"].value_counts().reset_index()

# Renombrar columnas para claridad
ventas_por_categoria.columns = ["Categoría del Producto", "Ventas"]

# Mostrar el resultado
display(ventas_por_categoria)
```

```python
# Contar cuántas veces se vendió cada categoría por tienda
ventas_categoria_tienda = df_completo.groupby(["Tienda", "Categoría del Producto"]).size().reset_index(name="Ventas")

# Mostrar la tabla
display(ventas_categoria_tienda)
```

```python
import matplotlib.pyplot as plt

# Agrupamos por tienda y categoría para contar las ventas
ventas_categoria_tienda = df_completo.groupby(["Tienda", "Categoría del Producto"]).size().reset_index(name="Ventas")

# Obtener la lista de tiendas
tiendas = ventas_categoria_tienda["Tienda"].unique()

# Crear un gráfico para cada tienda
for tienda in tiendas:
    # Filtrar los datos de esa tienda
    datos_tienda = ventas_categoria_tienda[ventas_categoria_tienda["Tienda"] == tienda]

    # Crear gráfico de barras
    plt.figure(figsize=(8, 5))
    plt.bar(datos_tienda["Categoría del Producto"], datos_tienda["Ventas"], color='coral')
    plt.title(f"Ventas por Categoría - {tienda}")
    plt.xlabel("Categoría del Producto")
    plt.ylabel("Cantidad de Ventas")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Agrupar por tienda y calcular la calificación promedio
calificacion_promedio = df_completo.groupby("Tienda")["Calificación"].mean().reset_index()

# Redondear a 2 decimales para mejor presentación
calificacion_promedio["Calificación"] = calificacion_promedio["Calificación"].round(2)

# Mostrar resultados
display(calificacion_promedio)
```

```python
# Crear gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(calificacion_promedio["Tienda"], calificacion_promedio["Calificación"], color='orange')

# Títulos y etiquetas
plt.title("Calificación Promedio por Tienda", fontsize=14)
plt.xlabel("Tienda", fontsize=12)
plt.ylabel("Calificación Promedio", fontsize=12)
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Contar cuántas veces se vendió cada producto
productos_mas_vendidos = df_completo["Producto"].value_counts().reset_index()

# Renombrar columnas
productos_mas_vendidos.columns = ["Producto", "Ventas"]

# Mostrar los 10 más vendidos
top_10_productos = productos_mas_vendidos.head(10)
display(top_10_productos)
```

```python
# Gráfico de los 10 productos más vendidos
plt.figure(figsize=(10, 6))
plt.bar(top_10_productos["Producto"], top_10_productos["Ventas"], color='steelblue')
plt.title("Top 10 Productos Más Vendidos", fontsize=14)
plt.xlabel("Producto", fontsize=12)
plt.ylabel("Cantidad de Ventas", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Mostrar los 10 productos menos vendidos
bottom_10_productos = productos_mas_vendidos.tail(10)
display(bottom_10_productos)
```

```python
# Gráfico de los 10 productos menos vendidos
plt.figure(figsize=(10, 6))
plt.bar(bottom_10_productos["Producto"], bottom_10_productos["Ventas"], color='crimson')
plt.title("10 Productos Menos Vendidos", fontsize=14)
plt.xlabel("Producto", fontsize=12)
plt.ylabel("Cantidad de Ventas", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Calcular el tiempo promedio de envío por tienda
tiempo_envio_promedio = df_completo.groupby("Tienda")["Cantidad de cuotas"].mean().reset_index()

# Renombrar la columna para claridad (opcional)
tiempo_envio_promedio.rename(columns={"Cantidad de cuotas": "Tiempo de Envío Promedio"}, inplace=True)

# Redondear a 2 decimales
tiempo_envio_promedio["Tiempo de Envío Promedio"] = tiempo_envio_promedio["Tiempo de Envío Promedio"].round(2)

# Mostrar la tabla
display(tiempo_envio_promedio)
```

```python
# Gráfico de barras del tiempo de envío promedio
plt.figure(figsize=(8, 5))
plt.bar(tiempo_envio_promedio["Tienda"], tiempo_envio_promedio["Tiempo de Envío Promedio"], color='purple')
plt.title("Tiempo Promedio de Envío por Tienda", fontsize=14)
plt.xlabel("Tienda")
plt.ylabel("Días Promedio de Envío")
plt.ylim(0, tiempo_envio_promedio["Tiempo de Envío Promedio"].max() + 1)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
